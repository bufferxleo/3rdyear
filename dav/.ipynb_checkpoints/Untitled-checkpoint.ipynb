```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
```

```python
y=iris.target
```

```python
df=pd.DataFrame(x)
```

```python
df['target']=iris.target
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,x,Y,y=train_test_split(df.drop('target',axis=1),df.target,test_size=0.2)

df_train=pd.DataFrame(X)
df_train['target']=Y
print(df_train)

df_test=pd.DataFrame(x)
df_test['target']=y
df_test
```

```python
l=[]
```

```python
for i in df.target:
    if i==0:
        l.append('setosa')
    elif i==1:
        l.append('versicolor')
    else:
        l.append('virginica')
```

```python
df['named_target']=l
```

```python
df
```

```python
X,x,Y,y=train_test_split(df.drop('target',axis=1),df.target,test_size=0.3)

df_train=pd.DataFrame(X)
df_train['target']=Y
print(df_train)

df_test=pd.DataFrame(x)
df_test['target']=y
print(df_test)
```

```python
d={'setosa':0,'versicolor':1,'virginica':2}
l=[]
for i in df.named_target:
    l.append(d[i])
```
