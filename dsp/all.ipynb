```python
#8
import pandas as pd
import numpy as np
def remove_outliers(col):
    q1=col.quantile(0.25)
    q3=col.quantile(0.75)
    iqr=q3-q1
    h=q1-iqr*1.5
    l=q3+iqr*1.5
    return h,l

from sklearn.datasets import load_iris
iris=load_iris()
col=pd.Series(iris.data[0])
h,l=remove_outliers(col)
col=np.where(col<l,l,col)
col=np.where(col>h,h,col)

col
```

```python
#10
from geopy.geocoders import Nominatim
def find_loc(lat,long):
    geolocator=Nominatim(user_agent="hi")
    loc=geolocator.reverse((lat,long),language='en')
    return loc
find_loc(34,100)[0].split(',')[:2]
```

```python
#11
from geopy.geocoders import Nominatim
def find_country(state):
    geolocator=Nominatim(user_agent='hi')
    c=geolocator.geocode(state)
    return c
find_country("goa")
```

```python
#13
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data)
df.columns=iris.feature_names
df['species']=iris.target

df

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


fig=plt.figure(figsize=(10,5))


plt.subplot(2,2,1)
sns.boxplot(y="sepal length (cm)",x='species',data=df)

plt.subplot(2,2,2)
sns.boxplot(y="petal length (cm)",x='species',data=df)

plt.subplot(2,2,3)
sns.boxplot(y="sepal width (cm)",x='species',data=df)

plt.subplot(2,2,4)
sns.boxplot(y="petal width (cm)",x='species',data=df)

plt.tight_layout()
```
