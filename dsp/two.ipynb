```python
import pandas as pd
# df=pd.read_csv("")

from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data)
df.columns=iris.feature_names
df['species']=iris.target

print(df)
print(df.describe())
```

```python
import numpy as np

corr=np.corrcoef(df)
corr
```

```python
import seaborn as sns
sns.heatmap(corr)
```
