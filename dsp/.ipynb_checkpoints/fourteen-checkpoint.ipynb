```python
from sklearn.datasets import load_iris
iris=load_iris()
import pandas as pd
import seaborn as sns

x=iris.data
y=iris.target

df=pd.DataFrame(x)
df.columns=iris.feature_names
df['species']=y
df=df[df.species==0]
df=df.iloc[:,0:2]
# print(df)
sns.jointplot(x='sepal length (cm)',y='sepal width (cm)',data=df,kind='kde')
```
