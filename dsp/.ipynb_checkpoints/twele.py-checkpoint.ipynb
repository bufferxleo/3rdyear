```python
from sklearn.datasets import load_iris
iris=load_iris()
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

x=iris.data
y=iris.target

df=pd.DataFrame(x)
df.columns=iris.feature_names
df['species']=y
# plt.jointplot
df=df[df.species==0]
sns.jointplot((df['sepal length (cm)'],df['sepal width (cm)']))
```
